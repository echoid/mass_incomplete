```python
import numpy as np

import numpy as np


# Define the file paths
train_data_path = 'data/train_data.txt'
test_data_path = 'data/test_data.txt'

train_label_path = 'data/train_labels.txt'
test_label_path = 'data/test_labels.txt'

# Load the data from the text files into NumPy arrays, handling 'nan' values
train_data = np.genfromtxt(train_data_path, delimiter=',', skip_header=0, dtype=float, filling_values=np.nan)
test_data = np.genfromtxt(test_data_path, delimiter=',', skip_header=0, dtype=float, filling_values=np.nan)

# Load the labels from the text files into NumPy arrays
train_labels = np.genfromtxt(train_label_path, delimiter=',', skip_header=0, dtype=float, filling_values=np.nan)
test_labels = np.genfromtxt(test_label_path, delimiter=',', skip_header=0, dtype=float, filling_values=np.nan)

# Combine the train and test data
combined_data = np.vstack((train_data, test_data))

# Combine the train and test labels
combined_labels = np.hstack((train_labels, test_labels))

# Save the combined data and labels to .npy files
np.save('data/rbf_feature.npy', combined_data)
np.save('data/rbf_labels.npy', combined_labels)
```

```python
# fetch dataset 
qsar_biodegradation = pd.read_csv("data/qsar_biodegradation/qsar_biodegradation.csv",sep=";")


# Assume the labels are in the last column and features are all other columns
features = qsar_biodegradation.iloc[:, :-1]
labels = qsar_biodegradation.iloc[:, -1]

#Apply label encoding to the labels
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Convert features and labels to NumPy arrays
features = np.array(features)
labels = np.array(encoded_labels)

# Print the features and labels to verify
print("Features:")
print(features)
print("Labels:")
print(labels)

# Save the features and labels to separate .npy files
np.save('data/qsar_biodegradation/feature.npy', features)
np.save('data/qsar_biodegradation/label.npy', labels)
```
